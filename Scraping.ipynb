```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
```

```python
path = "C:/Users/apriliawidyas/Documents/PROJECT MAGANG/chromedriver.exe"
driver = webdriver.Chrome(executable_path = path)
url = 'http://bpjt.pu.go.id/cek-tarif-tol'
driver.get(url)
```

```python
driver.find_elements_by_xpath('//*[@id="toll_road"]/option')
```

```python
road = []
gate1 = []
gate2 = []
gol = []
tarif = []

for i in range(len(driver.find_elements_by_xpath("//select[@name='troad']/option"))):
    driver.find_elements_by_xpath("//select[@class='soflow']/option")[i].click()
#     a = driver.find_elements_by_xpath("//select[@class='soflow']/option")[i].text
#     road.append(a)
    
    for j in range(len(driver.find_elements_by_xpath("//select[@name='tgates']/option"))):
        driver.find_elements_by_xpath("//select[@name='tgates']/option")[j].click()
#         a = driver.find_elements_by_xpath("//select[@name='tgates']/option")[j].text
#         gate1.append(a)
        
        for k in range(len(driver.find_elements_by_xpath("//select[@name='tgates2']/option"))):
            driver.find_elements_by_xpath("//select[@name='tgates2']/option")[k].click()
#             a = driver.find_elements_by_xpath("//select[@name='tgates2']/option")[k].text
#             gate2.append(a)
            
            for l in range(len(driver.find_elements_by_xpath("//select[@name='tgol']/option"))):
                driver.find_elements_by_xpath("//select[@name='tgol']/option")[l].click()
                
                a = driver.find_elements_by_xpath("//select[@class='soflow']/option")[i].text
                road.append(a)
                b = driver.find_elements_by_xpath("//select[@name='tgates']/option")[j].text
                gate1.append(b)
                c = driver.find_elements_by_xpath("//select[@name='tgates2']/option")[k].text
                gate2.append(c)
                d = driver.find_elements_by_xpath("//select[@name='tgol']/option")[l].text
                gol.append(d)
                
                soup=BeautifulSoup(driver.page_source, 'lxml')
                e = soup.find('p',class_='tarif').text
                tarif.append(e)
                
                time.sleep(1)
```

```python
import numpy as np
import pandas as pd

tol = {'Jalur' : road,
      'Pintu Masuk' : gate1,
      'Pintu Keluar' : gate2,
      'Kendaraan' : gol,
      'Tarif' : tarif}
```

```python
tol = pd.DataFrame(tol)
tol
```
